# District

```python
district = district.rename(columns={'A1':'district_id', 'A2':'district_name', 'A3':'region', 'A4':'total_inhabitants', 
                                    'A5':'num_munipalities_less_499','A6':'num_munipalities_500_to_1999', 
                                    'A7':'num_munipalities_2000_to_9999','A8':'num_munipalities_greater_10000','A9':'Total_cities', 
                                    'A10':'ratio_urban_inhabitants', 'A11':'average_salary', 'A12':'unemployment_rate_95', 
                                    'A13': 'unemployment_rate_96',
                                    'A14':'enterpreneurs_per_1000_inhabitants', 'A15':'num_crimes_commited_95', 'A16':'num_crimes_commited_96'})
```

```python
#While looking into unique values of umemployment rate column there is a discrepancy found, there are no null values but "?" in two columns
#which should be removed and filled with nan value (easy to find out the missing values)
#Third step is to replace those missing values with the mean value: group by region, as region is known for every value

# Replacing improper values "?" with missing values
district.unemployment_rate_95 = district.unemployment_rate_95.apply(lambda x: np.nan if x == "?" else x)
district.num_crimes_commited_95 = district.num_crimes_commited_95.apply(lambda x: np.nan if x == "?" else x)

# Converting the columns from object to float
district.unemployment_rate_95 = district.unemployment_rate_95.apply(float)
district.num_crimes_commited_95 = district.num_crimes_commited_95.apply(float)

# Dealing with missing values (replace with mean of the region)
district.unemployment_rate_95 = district.groupby("region").unemployment_rate_95.transform(lambda x: x.fillna(x.mean()))
district.num_crimes_commited_95 = district.groupby("region").num_crimes_commited_95.transform(lambda x: x.fillna(x.mean()))
```

```python
# Rename transaction columns
trans = trans.rename(columns={'amount': 'trans_amount', 'balance':'balance_after_trans', 
                                    'bank':'trans_bank_partner', 'account':'trans_account_partner',
                                    'type':'trans_type','operation':'trans_operation',
                                    'k_symbol':'trans_k_symbol'})
```

```python
##Change Date Format of trans date
trans['trans_date'] = pd.to_datetime(trans['date'], format = "%y%m%d", yearfirst = True)

#Delete date_account
trans = trans.drop(columns = "date")
```

```python
#lamnbda function is used as it is straightway assign an unknown value to the values that are not found in the table.
#Lambda function is powerful than normal function as it is reducing multiple lines of code and complexity

# convert transaction type column values to meaningful value
trans_type_lambda = defaultdict(lambda: 'UNKNOWN', 
                           {'PRIJEM': 'CREDIT', 
                            'VYDAJ': 'WITHDRAWAL'})

trans['trans_type'] = trans.trans_type.apply(lambda x: trans_type_lambda[x])

```

```python
#Similarly Applying lambda function to trans table:

# convert transaction operation column values to meaningful value, rename and missing and non matching values set to UNKNOWN
trans_operation_lambda = defaultdict(lambda: 'UNKNOWN', 
                           {'VYBER KARTOU': 'CC_WITHDRAWAL', 
                            'VKLAD': 'CREDIT_IN_CASH',
                            'PREVOD Z UCTU':'COLLECTION_FROM_OTHER_BANK',
                            'VYBER':'WITHDRAWAL_IN_CASH',
                            'PREVOD NA UCET':'REMITTANCE_TO_OTHER_BANK'})

trans['trans_operation'] = trans.trans_operation.apply(lambda x: trans_operation_lambda[x])
```

```python

# convert transaction K symbol-categorization column values to meaningful value
trans_k_symbol_lambda = defaultdict(lambda: 'UNKNOWN', 
                           {'POJISTNE': 'INSURANCE_PAYMENT', 
                            'SLUZBY': 'PAYMENT_FOR_STATEMENT',
                            'UROK': 'INTEREST_CREDITED',
                            'SANKC. UROK': 'SANCTION_INTEREST',
                            'SIPO': 'HOUSEHOLD',
                            'DUCHOD': 'OLD_AGE_PENSION',
                            'UVER': 'LOAN_PAYMENT'})

trans['trans_k_symbol'] = trans.trans_k_symbol.apply(lambda x: trans_k_symbol_lambda[x])
```

```python
#updating values of transtype to WITHDRAWL where the transtype is UNKNOWN
#so below statement update for WITHDRAWAL_IN_CASH trans operation
trans.loc[trans['trans_type'] == 'UNKNOWN', 'trans_type'] = 'WITHDRAWAL'
```

```python
#Considering the RFM Factor that is Recency, Frequency and Monetary
#in trans table the minimum and maximum value of transaction date is 1993-01-01 and 1998-12-31 that is six years interval

mean_balance_after_trans =trans.groupby(['account_id'],as_index=False)['balance_after_trans'].mean()
max_balance_after_trans =trans.groupby(['account_id'],as_index=False)['balance_after_trans'].max()
min_balance_after_trans =trans.groupby(['account_id'],as_index=False)['balance_after_trans'].min()


mean_trans_amount =trans.groupby(['account_id'],as_index=False)['trans_amount'].mean()
max_trans_amount =trans.groupby(['account_id'],as_index=False)['trans_amount'].max()
min_trans_amount =trans.groupby(['account_id'],as_index=False)['trans_amount'].min()


mean_balance_after_trans=mean_balance_after_trans.rename(columns= {'balance_after_trans': 'mean_balance_after_trans'})
max_balance_after_trans=max_balance_after_trans.rename(columns= {'balance_after_trans': 'max_balance_after_trans'})
min_balance_after_trans=min_balance_after_trans.rename(columns= {'balance_after_trans': 'min_balance_after_trans'})

mean_trans_amount=mean_trans_amount.rename(columns= {'trans_amount': 'mean_trans_amount'})
max_trans_amount=max_trans_amount.rename(columns= {'trans_amount': 'max_trans_amount'})
min_trans_amount=min_trans_amount.rename(columns= {'trans_amount': 'min_trans_amount'})

transaction_frequency =trans.groupby(['account_id'],as_index=False)['trans_id'].count()
transaction_frequency=transaction_frequency.rename(columns= {'trans_id': 'freq_total_trans'})

trans_freq=trans.groupby(['account_id','trans_date_recency_status']).size().rename('count').reset_index()
freq_first_3_years_trans=trans_freq.loc[trans_freq['trans_date_recency_status'] == 'First_3_years_trans']
freq_last_3_years_trans=trans_freq.loc[trans_freq['trans_date_recency_status'] == 'Last_3_years_trans']

freq_first_3_years_trans=freq_first_3_years_trans.rename(columns= {'count': 'freq_first_3_years_trans'})
freq_first_3_years_trans=freq_first_3_years_trans.rename(columns= {'count': 'freq_last_3_years_trans'})

credit=trans.groupby(['account_id','trans_type']).size().rename('count').reset_index()
freq_trans_credit= credit.loc[credit['trans_type'] == 'CREDIT']
freq_trans_withdrawl= credit.loc[credit['trans_type'] == 'WITHDRAWAL']

freq_trans_credit=freq_trans_credit.rename(columns= {'count': 'freq_credit_trans'})
freq_trans_withdrawl=freq_trans_withdrawl.rename(columns= {'count': 'freq_withdrawl_trans'})



max_date =trans.groupby(['account_id'],as_index=False)['trans_date'].max()
trans_freq_date = datetime.datetime(year=1999, day=1, month=1)
max_date['latest_trans_days']= trans_freq_date-max_date[['trans_date']]
```

```python
trans_rfm=pd.merge(mean_balance_after_trans, max_balance_after_trans, on='account_id', how='inner')
trans_rfm=pd.merge(trans_rfm, transaction_frequency, on='account_id', how='inner')
trans_rfm=pd.merge(trans_rfm, mean_trans_amount, on='account_id', how='inner')
trans_rfm=pd.merge(trans_rfm, max_trans_amount, on='account_id', how='inner')
trans_rfm=pd.merge(trans_rfm, min_trans_amount, on='account_id', how='inner')
trans_rfm=pd.merge(trans_rfm, freq_first_3_years_trans, on='account_id', how='inner')
trans_rfm=pd.merge(trans_rfm, freq_last_3_years_trans, on='account_id', how='inner')
trans_rfm=pd.merge(trans_rfm, freq_trans_credit, on='account_id', how='inner')
trans_rfm=pd.merge(trans_rfm, freq_trans_withdrawl, on='account_id', how='inner')
trans_rfm=pd.merge(trans_rfm, max_date, on='account_id', how='inner')
```
